### Customizing plots

```python
df2.plot()
```

```python
df2['c'].plot.line(figsize=(20,10),ls='--')
```

```python
df2['c'].plot.line(figsize=(10,3),ls=':',c='green',lw=5)
```

```python
title='MY PLOT TITLE'
xlabel= 'MY X DATA'
ylabel='MY Y DATA'
```

```python
ax = df2['c'].plot.line(figsize=(10,3),ls=':',c='green',lw=5,title=title)
ax.set(xlabel=xlabel,ylabel=ylabel)
```

```python
ax=df2.plot()
ax.legend(loc=9)
```

```python
ax=df2.plot()
ax.legend(loc=9,bbox_to_anchor=(1.1,1))
```
